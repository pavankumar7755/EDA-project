### Q7. Determine what percentage of churned customers are aged over 45. Compare it to their overall percentage in the dataset.

```python
# this will Return the value of total churned data 
total_churn_customer= df.Exited.sum()  
# this will Return the value of over 45  churned data 
total_churn_customer_over_45=df[df['Age']>45].Exited.sum()

Percentage_over45=(total_churn_customer_over_45/total_churn_customer)*100    # 
print(f"Total count of Churned customer :{Percentage_over45 :.2f}%")

Percentage_overall_over45=(total_churn_customer_over_45/df.shape[0])*100
print(f"overall percentage of over 45 :{Percentage_overall_over45:.2f}%")



##  alterbative syntax 
total_customers_over_45 = df[df['Age'] > 45].shape[0]
churn_rate_over_45 = (total_churn_customer_over_45 / total_customers_over_45) * 100
print(f"Churn rate among customers over 45: {churn_rate_over_45:.2f}%")
```

###  Q9. Analyze the churn rate based on NumOfProducts. Create a summary table showing churn rate for 1, 2, 3, and 4 products.

```python
numofproduct=df.groupby('NumOfProducts').Exited.agg(Sum='sum',Mean='mean',count='count')
numofproduct['Mean']=numofproduct['Mean']*100    # convert mean to percenatge
numofproduct
```

```python
# Group by NumOfProducts and calculate churn rate
summary_table = df.groupby('NumOfProducts')['Exited'].agg(
    Total_Customers='count',
    Churned_Customers='sum',
    Churn_Rate=lambda x: round(x.mean() * 100, 2)  # Percentage
).reset_index()

print(summary_table)
```
